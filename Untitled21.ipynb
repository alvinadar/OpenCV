```python
# sir i have used ultralytics library instead of haar cascade to detect animal face
import cv2
from ultralytics import YOLO
from google.colab.patches import cv2_imshow

# This loads the image of the frog
image = cv2.imread('/content/frog.jpeg')
small_image = cv2.resize(image, (300, 300))

# loads the YOLO library
model = YOLO('yolov8n.pt')
#Runs the model
results = model.predict(small_image)

for result in results:
    for box in result.boxes:
        cls = int(box.cls[0])
        label = model.names[cls]
        x1, y1, x2, y2 = map(int, box.xyxy[0])

        # this line of the code below is to draw the green rectangle on the object detected
        cv2.rectangle(small_image, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(small_image, label, (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)



cv2_imshow(small_image)

```
